# SVM for classification

Load IRIS dataset, check its contents:

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris.feature_names
print(iris.feature_names)
print(iris.data[0:5,:])
print(iris.target[0:5])
#print(iris.data)
print(iris.target) # Labels of flowers (0,1,2)
print(iris.target_names) #Different types of flowers in the dataset
```

Split data into training and testing parts:

```python
from sklearn.model_selection import train_test_split
X=iris.data # each row represents a flower, and each column its features
y=iris.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Use a Support Vector Machine for classification:

```python
from sklearn.svm import SVC
SVMmodel=SVC(kernel='linear') # Linear kernel: draws a straight line to separate classes
SVMmodel.fit(X_train,y_train)
SVMmodel.get_params()
SVMmodel.score(X_test,y_test)
```

Let's explore more now.

Choose only first two features (columns) of iris.data
SVM is in its basic form a 2-class classifier, so eliminate iris.target =2 from the data


Plot scatterplots of targets 0 and 1 and check the separability of the classes:



```python
import matplotlib.pyplot as plt

plt.scatter(X[y == 0, 0], X[y == 0, 1], label="Setosa", color="blue")
plt.scatter(X[y==1,0], X[y==1,1], label='Versicolor', color='red')
plt.scatter(X[y==2,0], X[y==2,1], label='Virginica', color='green')

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.title("Scatter plot of Setosa, Versicolor and Virginica")
plt.show()
```

Train and test the SVM classifier, play with regularization parameter C (either use the default value or try e.g. 200)



```python
X = iris.data[iris.target!=2,0:2]
y = iris.target[iris.target!=2]
SVMmodel1_1 = SVC(kernel='linear')
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X,y,test_size=0.2)
SVMmodel1_1.fit(X_train_1,y_train_1)
SVMmodel1_1.score(X_test_1,y_test_1)
```

Show support vectors in the 2D plot, plot the decision line from equation [w0 w1]*[x0 x1] + b = 0:

```python
import numpy as np
import matplotlib.pyplot as plt

supvectors=SVMmodel1_1.support_vectors_
# Plot the support vectors here
print(supvectors)
plt.scatter(supvectors[:,0], supvectors[:,1], s=100, facecolors='none', edgecolors='black', label='Support Vectors')

#Separating line coefficients:
W=SVMmodel1_1.coef_
b=SVMmodel1_1.intercept_
print(W)
print(b)
#The equation for linear Kernel SVM is ->
#w1*x1+w2*x2+b=0
#x2=-(b+w1*x1)/w2

x1 = np.linspace(np.min(X[:,0]),np.max(X[:,0]),100)
print(x1.shape, W.shape, b.shape)
x2=-(b+W[0,0]*x1)/W[0,1]
plt.plot(x1, x2, color='red', label='Split Line')
plt.scatter(X[y==0,0], X[y==0,1], color='blue', label='Class 0')
plt.scatter(X[y==1,0], X[y==1,1], color='orange', label='Class 1')
plt.legend()
```

# Anomaly detection via SVM

Import one-class SVM and generate data (Gaussian blobs in 2D-plane)

```python
from sklearn.svm import OneClassSVM
from sklearn.datasets import make_blobs
from numpy import quantile, where, random

random.seed(11) # Set a random seed for reproducibility
x, _ = make_blobs(n_samples=300, centers=1, cluster_std=.3, center_box=(4, 4)) # Generate a dataset with 300 samples forming a single cluster

plt.scatter(x[:,0], x[:,1])
plt.show()
```

Train one-class SVM and plot the outliers (outputs of prediction being equal to -1)

```python
SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.03)
#The 'rbf' kernel is commonly used because it can capture complex decision boundaries.
#Gamma: A higher game value means that each training point has a shorter influence radius,
#leading to a more complex decision boundary. A lower value makes the model more general.

SVMmodelOne.fit(x)
pred = SVMmodelOne.predict(x)
anom_index = where(pred==-1)
values = x[anom_index]

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()
```


Plot the support vectors:

```python
supvectors1=SVMmodelOne.support_vectors_
# Plot the support vectors here
print(supvectors1)
plt.scatter(supvectors1[:,0], supvectors1[:,1], s=100, facecolors='none', edgecolors='black', label='Support Vectors')
plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()
```

What if we want to have a control what is outlier? Use e.g. 5% "quantile" to mark the outliers. Every point with lower score than threshold will be an outlier.

```python
scores = SVMmodelOne.score_samples(x)
print(scores)
thresh = quantile(scores, 0.01)
print(thresh)
index = where(scores<=thresh)
values = x[index]

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()
```
